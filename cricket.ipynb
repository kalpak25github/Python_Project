```python
from bs4 import BeautifulSoup
import requests
```

```python
webpage=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/t20i")
```

```python
soup=BeautifulSoup(webpage.content)
print(soup.prettify())
```

```python
tabledata=soup.find_all("table")
tabledata
```

```python
alldata=[]

for i in tabledata:
    for j in i.find_all("td"):
        alldata.append(j.text.replace("\n",""))
alldata    
```

```python
pos=alldata[0:len(alldata):5]
pos
```

```python
teams=alldata[1:len(alldata):5]
teams
```

```python
matches=alldata[2:len(alldata):5]
matches
```

```python
points=alldata[3:len(alldata):5]
points
```

```python
ratings=alldata[4:len(alldata):5]
ratings
```

```python
first=ratings[0]
first=first.strip()
first
```

```python
ratings[0]=first
ratings
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df["Position"]=pos
df["Teams"]=teams
df["Matches"]=matches
df["Points"]=points
df["Ratings"]=ratings
df
```

```python
df.to_csv("rankingsinfo")
```

```python
import pymysql as ps
```

```python
mycon=ps.connect(host="localhost",user="root",database="cricket",password="")
mycur=mycon.cursor()
mycur.execute("create table if not exists rankingsinfo(Position int,Teams varchar(20),Matches int,Points int,Ratings int)")
```

```python
for i in range(len(pos)):
    mycur.execute("""insert into rankingsinfo values('%s','%s','%s','%s','%s')"""%(pos[i],teams[i],matches[i],points[i],ratings[i]))    
    mycon.commit()

```
